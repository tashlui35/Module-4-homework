# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
# Reading whale returns
whale_path = Path("Resources/whale_returns.csv")
whale_df = pd.read_csv(whale_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
# Count nulls
whale_df.isnull().mean() * 100
```

```python
# Drop nulls
whale_df = whale_df.dropna().copy()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_path = Path("Resources/algo_returns.csv")
algo_df = pd.read_csv(algo_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
# Count nulls
algo_df.isnull().mean() * 100
```

```python
# Drop nulls
algo_df = algo_df.dropna().copy()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sp_path = Path("Resources/sp_tsx_history.csv")
sp_df = pd.read_csv(sp_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
# Check Data Types
sp_df.dtypes
```

```python
# Fix Data Types
sp_df['Close'] = sp_df['Close'].astype('str').str.replace(',', '').str.replace('$', '').astype('float')
```

```python
sp_df.head()
```

```python
sp_df.dtypes
```

```python
# Calculate Daily Returns
daily_returns = sp_df.pct_change().copy()
```

```python
# Drop nulls
daily_returns = daily_returns.dropna().copy()
```

```python
# Rename `Close` Column to be specific to this portfolio.
daily_returns.columns = ["S&P TSX 60"]
daily_returns.head()
```

```python
whale_df.dtypes
```

```python
algo_df.dtypes
```

```python
sp_df.dtypes
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
combined = pd.concat( [whale_df, algo_df, daily_returns], axis="columns", join="inner" )
```

```python
combined.head()
```

# Conduct Quantitative Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance Anlysis

#### Calculate and Plot the daily returns.

```python
# Plot daily returns of all portfolios
combined.plot(figsize=(15, 8))
```

#### Calculate and Plot cumulative returns.

```python
# Calculate cumulative returns of all portfolios
cumulative_returns = (1 + combined).cumprod() - 1
# Plot cumulative returns
cumulative_returns.plot(figsize=(15, 8))
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios.
4. Determine which portfolios are riskier than the S&P TSX 60.
5. Calculate the Annualized Standard Deviation.

### Create a box plot for each portfolio


```python
# Box plot to visually show risk
combined.boxplot(figsize=(15, 8))
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
daily_std = combined.std()
```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculate  the daily standard deviation of S&P TSX 60
sp_std = daily_returns["S&P TSX 60"].std()
# Determine which portfolios are riskier than the S&P TSX 60
riskier_portfolios = daily_std[daily_std > sp_std]
riskier_portfolios
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = combined.std() * np.sqrt(252)
annualized_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rolling_std = combined.rolling(window=21).std()
# Plot the rolling standard deviation
rolling_std.plot(figsize=(10,5), title="Rolling Standard Deviation (21-day window)")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = combined.corr()
# Display de correlation matrix
sns.heatmap(correlation)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
covariance = combined['BERKSHIRE HATHAWAY INC'].cov(combined['S&P TSX 60'])
# Calculate variance of S&P TSX
variance = combined['S&P TSX 60'].var()
# Computing beta
beta = covariance / variance
# Plot beta trend
rolling_beta = combined['BERKSHIRE HATHAWAY INC'].rolling(window=21).cov(combined['S&P TSX 60']) / combined['S&P TSX 60'].rolling(window=21).var()
rolling_beta.plot(figsize=(10,5), title="Rolling Beta (21-day window)")
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
ewm_std = combined.ewm(span=21).std()
ewm_std
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratio = (combined.mean() / combined.std()) * np.sqrt(252)
annualized_sharpe_ratio = sharpe_ratio * np.sqrt(252)
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratio.plot(kind="bar", figsize=(10,5), title="Sharpe Ratio")
```

```python
annualized_sharpe_ratio.plot(kind="bar", figsize=(10,5), title="Annualized Sharpe Ratio")
```

To determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios, we need to analyze the returns and risk metrics of each portfolio.

Firstly, we analyzed the cumulative returns of all portfolios. From the analysis, we can see that the algorithmic portfolios outperformed the whales portfolios but underperformed the S&P TSX 60 market index. Specifically, the Algo 1 portfolio had the highest cumulative return, followed by BERKSHIRE, thenAlgo 2. SOROS, PAULSON had lower cumulative returns than the algorithmic portfolios but still outperformed the S&P TSX 60 index.

Next, we looked at the daily standard deviation of all portfolios. We found that the algorithmic portfolios had higher daily standard deviations than the market index but lower than some of the whale portfolios. This indicates that the algorithmic portfolios have higher risk than the market but lower risk than some of the whale portfolios.

We also calculated the annualized standard deviation of all portfolios, which takes into account the number of trading days in a year (252). We found that the algorithmic portfolios had higher annualized standard deviations than the market index but lower than some of the whale portfolios.

Additionally, we analyzed the rolling standard deviation of all portfolios using a 21-day window. The analysis showed that the algorithmic portfolios had higher rolling standard deviations than the market index but lower than some of the whale portfolios.

Furthermore, we calculated the Sharpe ratios of all portfolios. The Sharpe ratio is a measure of risk-adjusted returns, and a higher Sharpe ratio indicates a better risk-adjusted performance. We found that the algorithmic portfolios had higher Sharpe ratios than the market index but lower than some of the whale portfolios.

Based on the analysis, we can conclude that the algorithmic portfolios outperformed the whale portfolios in terms of cumulative returns and risk-adjusted performance. However, they underperformed the S&P TSX 60 market index. Therefore, it is difficult to say whether the algorithmic strategies outperform both the market and the whales portfolios, as it depends on the specific metric and time period analyzed. ## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Reading data from 1st stock
l_path = Path("Resources/l_historical.csv")
l_df = pd.read_csv(l_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
l_df = l_df.drop(columns=["Symbol"])
```

```python
# Reading data from 2nd stock
otex_path = Path("Resources/otex_historical.csv")
otex_df = pd.read_csv(otex_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
otex_df = otex_df.drop(columns=["Symbol"])
```

```python
# Reading data from 3rd stock
shop_path = Path("Resources/shop_historical.csv")
shop_df = pd.read_csv(shop_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
shop_df = shop_df.drop(columns=["Symbol"])
```

```python
# Combine all stocks in a single DataFrame
custom_combined = pd.concat([l_df, otex_df, shop_df], axis="columns", join="inner")
custom_combined.head()
```

```python
# Reset Date index
custom_combined.reset_index()
```

```python
# Reorganize portfolio data by having a column per symbol
custom_combined.columns = ["L", "OTEX", "SHOP"]
```

```python
# Calculate daily returns
custom_daily_returns = custom_combined.pct_change()
# Drop NAs
custom_daily_returns = custom_daily_returns.dropna()
# Display sample data
custom_daily_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
portfolio_returns = custom_daily_returns.dot(weights)
# Display sample data
portfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
custom_combined_returns = pd.concat([portfolio_returns, custom_daily_returns], axis="columns", join="inner")
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
custom_combined_returns = custom_combined_returns.dropna()
custom_combined_returns.columns = ["RETURNS","L", "OTEX", "SHOP"]
custom_combined_returns.head()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
custom_std = custom_combined_returns.std() * np.sqrt(252)
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
rolling_std = custom_combined_returns.rolling(window=21).std()
# Plot rolling standard deviation
rolling_std.plot(figsize=(10,5), title="Rolling 21-Day Standard Deviation")
```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation
correlation = custom_combined_returns.corr()
sns.heatmap(correlation, annot=True)
```

### Calculate and Plot the 60-day Rolling Beta for Your Portfolio compared to the S&P 60 TSX

```python
custom_combined_returns.head()
```

```python
# Calculate and plot Beta
covariance = custom_combined_returns['SHOP'].cov(custom_combined_returns['RETURNS'])
beta = covariance / variance
beta_plot = custom_combined_returns['SHOP'].rolling(window=21).cov(custom_combined_returns['RETURNS']) / variance
beta_plot.plot(figsize=(10,5), title="Rolling 21-Day Beta")
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualized Sharpe Ratios
annualized_returns = custom_combined_returns.mean() * 252
annualized_volatility = custom_combined_returns.std() * np.sqrt(252)
sharpe_ratios = annualized_returns / annualized_volatility
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind='bar', figsize=(10,5), title="Annualized Sharpe Ratios")
```
